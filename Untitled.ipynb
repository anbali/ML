```python
import codecs
import json
import os
import re
import sys
from skimage import io
from skimage import measure
from skimage.color import rgb2gray
from skimage.filters import roberts, threshold_otsu
from skimage.morphology import convex_hull_image, skeletonize
import matplotlib.pyplot as plt
```

```python
# Константы, полученные экспертным путем
threshold_value = 128 # значение для порогового преобразования
contours_level = 0.15 # значение для поиска контуров
delta = 30
path='/home/ds/anna/ЦОИ/Apple'
# Получаем список преобразованных для дальнейшей обработки изображений
def get_list_image(path):
    list_image = []
    for d, dirs, files in os.walk(path): # Достаем все каталоги по пути
        for f in files: # Достаем все файлы
            if re.search(".jpg", f): # Сортировка по типу
                path = os.path.join(d,f) # формирование адреса 
                picture = io.imread(path) # Загружаем изображение в формате narray
                picture_gray = rgb2gray(picture) # Переводим изображений в полутон
                picture_roberts = roberts(picture_gray) #применяем фильтр Робертса
                thresh = threshold_otsu(picture_roberts,threshold_value) # выполняем пороговое бинарное преобразование
                picture_binary = picture_roberts > thresh #выполняем пороговое бинарное преобразование
                chull = convex_hull_image(picture_binary) #Делаем обертку
                skeleton = skeletonize(chull) # Находим скилет
                io.imshow(skeleton)
                list_image.append(skeleton) # Добавляем полученное изображение в список
    return list_image
```

```python
# Получаем самые большие контуры, максимум 2
def get_size_contours(list_image):
    list_contours = []
    i = 1
    sum_len_contour = 0
    for picture in list_image: # Пробегаем по картинкам
        contours = measure.find_contours(picture, contours_level) # находим контуры
        len_contour = 0
        for contour in contours:
            len_contour += len(contour) # Размер контуров для  одной картинки
            sum_len_contour += len(contour) # Размер контуров от одной кортинки для всех
        list_contours.append(len_contour) # Добавляем в список, для последующего нахождения максимума
        i += 1    
    print(list_contours)   
    average = sum_len_contour / i # Находим среднее
    print(average)
    min_len_contour = 0
    if average - delta > 0:
        min_len_contour = average - delta # Вычисляем минимальную границу
    max_len_contour = max(list_contours) # Вычисляем максимальную границу
    delta_training = [min_len_contour, max_len_contour] #Получаем дельту для тренировки
    return delta_training
```

```python
def find_image(path, delta_training):
    min_len_contout = delta_training[0]
    max_len_contour = delta_training[1]
    finding_images = ""
    for d, dirs, files in os.walk(path): # Достаем все каталоги по пути
        for f in files: # Достаем все файлы
            if re.search(".jpg", f): # Сортировка по типу
                path = os.path.join(d, f) # Получаем путь до изображения
                picture = io.imread(path) # Загружаем изображение в формате narray
                picture_gray = rgb2gray(picture) # Переводим изображений в полутон
                picture_roberts = roberts(picture_gray) #применяем фильтр Робертса
                thresh = threshold_otsu(picture_roberts,threshold_value) # выполняем пороговое бинарное преобразование
                picture_binary = picture_roberts > thresh #выполняем пороговое бинарное преобразование
                chull = convex_hull_image(picture_binary) #Делаем обертку
                skeleton = skeletonize(chull) # Находим скилет
                contours = measure.find_contours(skeleton,contours_level) # находим контуры
                summ_len_contour = 0
                for contour in contours:
                    summ_len_contour += len(contour) # находим длину контура # Проверяем попадает ли длина контура взаданный диапазон
                    if (summ_len_contour >= min_len_contout) and (summ_len_contour <= max_len_contour):
                        finding_images += path + "\t true\n"
                    else:
                        finding_images += path + "\t false\n"
    return finding_images
```

```python
obj_text = codecs.open("/home/ds/anna/ЦОИ/Apple.json",'r', encoding='utf-8').read()
# Открываем файл с тренировочными данными
delta_training = json.loads(obj_text) # Загружаем средние
findingImages = find_image("/home/ds/anna/ЦОИ/Apple2",delta_training) # Ищем все совпадения сзаданными контурами
# записываем рузультам в файл
with open("resultTesting_Apple_Red_Delicious.cvs", "w") as file:
    print(findingImages, file=file)
```
